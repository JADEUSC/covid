```python
import jhu
import pandas as pd
import matplotlib.pyplot as plt
```

```python
US = jhu.load_us()

start = pd.Timestamp('2020-03-01')

US['tot'][start:].plot(style='o')
plt.title('US')
#plt.yscale('log')
plt.show()


US['MA'][start:].plot(style='o')
plt.title('MA')
#plt.yscale('log')
plt.show()

```

```python
for state in US.columns.unique(level=0):
    US[state][start:].plot(style='o')
    plt.title(state)
    plt.show()
```
